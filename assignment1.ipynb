# Visualisation

# V1: Histogram

```python
plt.figure()
plt.hist(df['SL'],width=0.2,color='green',cumulative=True)
plt.title('SL',fontsize=17)
plt.xticks()
plt.yticks()
```

```python
df[['SL','SW']].plot.kde()
plt.title('SL & SW',fontsize=17)
```

# V2:Count plot

```python
sns.countplot(df['PW'])
plt.title('PW',fontsize=17)
```

# V3: Box plot

```python
sns.boxplot(x='SW',y='PW',data=df)
plt.title('SL & SW',fontsize=17)
```
